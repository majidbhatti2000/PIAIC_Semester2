# Task 2

```python
import glob
import pandas as pd

files = glob.glob('states*.csv')

us_census = pd.concat([pd.read_csv(f) for f in glob.glob('states*.csv')])
us_census.head()
```

# Task 3

```python
us_census.columns
```

```python
us_census.dtypes
```

# Task 4

```python
us_census.head()
```

```python
us_census['Income'].dtype
us_census['TotalPop'] = us_census['TotalPop'].astype(int)
```

# Task 5

```python
us_census['Income'].dtype
us_census['Income'] = us_census.Income.str.replace('$','',regex=True)
```

```python
us_census['Income'] = us_census['Income'].astype(float)
us_census.dtypes
```

# Task 6

```python
us_census[['Male','Female']] = us_census.GenderPop.str.split('_',expand=True)
```

```python
us_census.head()
```

# Task 7

```python
us_census['Male'] = us_census.Male.str.replace('M','')
us_census['Female'] = us_census.Female.str.replace('F','')
```

```python
import numpy as np

us_census['Male'] = us_census['Male'].astype(int)
us_census['Female'] = us_census['Female'].replace('', np.nan).fillna(0).astype(int)
```

```python
us_census.dtypes
```

```python
us_census.head()
```

# Task 8

```python
import matplotlib.pyplot as plt

plt.scatter(us_census['Female'],us_census['Income'])
plt.xlabel('Female')
plt.ylabel('Income')
plt.show()
```

# Task 9

```python
us_census['Female'] = us_census['TotalPop'] - us_census['Male']
```

# Task 13

```python
us_census.dtypes
us_census["Hispanic"] = us_census.Hispanic.str.replace('%','',regex=True).astype(float)
us_census["White"] = us_census.White.str.replace('%','',regex=True).astype(float)
us_census["Black"] = us_census.Black.str.replace('%','',regex=True).astype(float)
us_census["Native"] = us_census.Native.str.replace('%','',regex=True).astype(float)
us_census["Asian"] = us_census.Asian.str.replace('%','',regex=True).astype(float)
us_census["Pacific"] = us_census.Pacific.str.replace('%','',regex=True).astype(float)
us_census.head()
```
